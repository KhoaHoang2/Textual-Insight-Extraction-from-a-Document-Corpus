```python
#Task 1: Load and Clean
with open('calculustext.txt', 'r') as f:
    content= f.read().lower().replace(',','').replace('.','') 
    content = ''.join([c for c in content if not c.isdigit() ])
    content = ' '.join(content.split())
content
```

```python
#Task 2: Split and Stop Word Removal
Content = content.split()
stop = ['The', 'a', 'is', 'in','to', 'it', 'and', 'that', 'for', 'with', 'on', 'at', 'from', 'he', 'she', 'they', 'was', 'were', 'had', 'have', 'this', 'or', 'but', 'not', 'as', 'by', 'which', 'what', 'when', 'where', 'how', 'all', 'any', 'most', 'other', 'your', 'our', 'my', 'me', 'you']

filter_token=[]
for c in Content:
    if c not in stop:
        filter_token.append(c)
print(filter_token)
```

```python
#Task 3: Create Vocabulary 
vocab = set(filter_token)
w_count = {}

#Calculate word frequencies
for w in Content:
    if w in w_count:
        w_count[w] += 1
    else:
        w_count[w] = 1

#Disply top 10 most frequent words
sort_d = sorted(w_count.items(),key = lambda x: x[1])
top_10 = sort_d[-10:]
print(top_10)
```

```python
#Task 4
#Bigram Generation
def bigram(L):
    token = tuple(zip(L,L[1:]))
    return token
bi= bigram(filter_token)
        
#Trigram Generation
def trigram(L):
    token = tuple(zip(L, L[1:], L[2:]))
    return token
tri = trigram(filter_token)

#N-gram Generation
def ngram(L):
    bi_gr,tri_gr = {},{}
#Bigram top 10
    for w in bi:
        if w in bi_gr:
            bi_gr[w] += 1
        else:
            bi_gr[w] = 1            
    sort_d = sorted(bi_gr.items(),key = lambda x: x[1])
    bi_top_10 = sort_d[-10:]
    
#Trigram top 10
    for w in tri:
        if w in tri_gr:
            tri_gr[w] += 1
        else:
            tri_gr[w] = 1
    sort_d = sorted(tri_gr.items(),key = lambda x: x[1])
    tri_top_10 = sort_d[-10:]
    return bi_top_10, tri_top_10
    
ngram(filter_token)
```
